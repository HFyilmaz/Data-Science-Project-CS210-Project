```python
num_comments = df.shape[0]
sentiment_average = df['sentiment_score'].mean()
kk_average = kk_df['sentiment_score'].mean()
rte_average = rte_df['sentiment_score'].mean()
so_average = so_df['sentiment_score'].mean()
mi_average = mi_df['sentiment_score'].mean()
kk_mentioned = kk_df.shape[0]
rte_mentioned = rte_df.shape[0]
so_mentioned = so_df.shape[0]
mi_mentioned = mi_df.shape[0]

print('number of comments', num_comments)
print('sentiment average:', sentiment_average)
print('kk mentioned:', kk_mentioned)
print('kk sentiment average:', kk_average)
print('rte mentioned:', rte_mentioned)
print('rte sentiment average:', rte_average)
print('so mentioned:', so_mentioned)
print('so sentiment average:', so_average)
print('mi mentioned:', mi_mentioned)
print('mi sentiment average:', mi_average)
```

```python
# butun commentlerin ay bazinda dagilimi
counts1 = df.groupby(df['publishDate'].dt.to_period('M')).size()
plt.figure(figsize=(10, 5))
counts1.plot(kind='bar')
plt.xlabel('Year and Month')
plt.ylabel('Frequency of comments in a certain year and month')
plt.title('Distribution of comments by Year and Month')
plt.show()
```

```python
count_feb = counts1['2023-02']
count_mar = counts1['2023-03']
count_apr = counts1['2023-04']
count_may = counts1['2023-05']

print('comments in feb:', count_feb)
print('comments in mar:', count_mar)
print('comments in apr:', count_apr)
print('comments in may:', count_may)
```
